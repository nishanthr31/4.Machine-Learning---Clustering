```python
import pandas as pd
dataset = pd.read_csv("Mall_Customers.csv")
dataset
```

```python
X= dataset.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import MeanShift
Ap = MeanShift(bandwidth=25,max_iter=700)
CN = Ap.fit_predict(X)
CN
```

```python
supervised = dataset
```

```python
supervised['Cluster_group'] = CN
supervised
```

```python
import seaborn as sb
graph = sb.lmplot(data=supervised,x=supervised.columns[3],y= supervised.columns[4],hue= supervised.columns[5],fit_reg=False,legend=True,legend_out= True)
```
